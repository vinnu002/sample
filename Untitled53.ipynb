```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the provided CSV string
data = pd.read_csv("your_data.csv")

# Replace '<1' and '-' with 0 for easier plotting
data.replace({'<1': 0, '-': 0}, inplace=True)

# Convert percentage columns to numeric
for col in data.columns[2:]:
    data[col] = pd.to_numeric(data[col])

# Example visualizations:

# 1. Trend of female representation over time for a specific company
company_name = "Google"  # Replace with the desired company
company_data = data[data["Company"] == company_name]
plt.figure(figsize=(10, 6))
plt.plot(company_data["Year"], company_data["Female %"], marker='o', linestyle='-')
plt.xlabel("Year")
plt.ylabel("Female Representation (%)")
plt.title(f"Trend of Female Representation at {company_name}")
plt.grid(True)
plt.show()

# 2. Comparison of racial/ethnic diversity across companies in a specific year
target_year = 2018  # Replace with the desired year
year_data = data[data["Year"] == target_year]
race_cols = ['% White', '% Asian', '% Latino', '% Black', '% Multi', '% Other']
year_data.set_index("Company")[race_cols].plot(kind="bar", stacked=True, figsize=(12, 6))
plt.xlabel("Company")
plt.ylabel("Racial/Ethnic Representation (%)")
plt.title(f"Racial/Ethnic Diversity across Companies in {target_year}")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# 3. Gender representation by job category for a specific company and year
company_name = "Apple"
target_year = 2016
company_year_data = data[(data["Company"] == company_name) & (data["Year"] == target_year)]


plt.title(f"Gender Representation by Job Category at {company_name} in {target_year}")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# Additional examples and customization:

# - Create box plots to show the distribution of diversity metrics across companies.
# - Calculate and visualize other diversity indices.
# - Use seaborn for more visually appealing and informative plots.
# - Experiment with different plot types and settings to find the best way to represent the data.
```
